```python
import pandas as pd
from sklearn.model_selection import train_test_split

# Read the CSV file
df = pd.read_csv('PonziDataset_20221114.csv')

# Extract features (X) and labels (Y)
X = df['address']  # Assuming 'label' is the column you want to predict
Y = df['label']

# Assuming you have normalized your data and assigned it to normalized_data_array
# If not, you may need to perform normalization on your features before splitting the data

# Perform train-test split
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.1, random_state=42, stratify=Y)

# Now, you can save the train and test datasets to separate CSV files
train_data = pd.concat([X_train, Y_train], axis=1)
test_data = pd.concat([X_test, Y_test], axis=1)

# Save to CSV files
train_data.to_csv('train_data.csv', index=False)
test_data.to_csv('test_data.csv', index=False)
```
